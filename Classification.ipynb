```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import jaccard_score, plot_confusion_matrix
from sklearn.preprocessing import StandardScaler
```

```python
df1 = pd.read_csv('train_set.csv')
df2 = pd.read_csv('train_NDVI_misra_b5.csv')
```

```python
df2 = df2.drop(columns=['years', 'crop_type'])
```

```python
df1
```

```python
df2
```

```python
df = pd.concat([df1, df2], axis=1)
df
```

```python
# df = df1
```

```python
X = df.loc[:, (df.columns != 'crop_type') & (df.columns != 'years')].to_numpy()
Y = df.crop_type.to_numpy()
X.shape, Y.shape
```

```python
seed = 7
test_size = 0.2
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=test_size, random_state=seed)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# clf1 = XGBClassifier(grow_policy='lossguide')
# clf1 = XGBClassifier(seed=seed, eta=0.02, n_estimators=225, max_depth=5, subsample=1, gamma=0.01, grow_policy='lossguide')
clf1 = XGBClassifier(seed=seed, eta=0.01, n_estimators=250, max_depth=5, subsample=1, gamma=0.01)
clf2 = RandomForestClassifier(n_estimators=50, random_state=seed)
clf3 = GaussianNB()
clf4 = LinearDiscriminantAnalysis()
clf5 = KNeighborsClassifier(n_neighbors=7)
# clf6 = MLPClassifier(random_state=seed, hidden_layer_sizes=(256,), max_iter=300)
eclf1 = VotingClassifier(estimators=[
    ('xgb', clf1), ('rf', clf2), ('gnb', clf3), ('lda', clf4), ('knn', clf5)#, ('mlp', clf6)
], voting='soft')
eclf1 = eclf1.fit(X_train_scaled, y_train)
```

```python
y_train_pred = eclf1.predict(X_train_scaled)
y_test_pred = eclf1.predict(X_test_scaled)
```

```python
# clf1.fit(X_train_scaled, y_train)
# y_test_pred = clf1.predict(X_test_scaled)
```

```python
# print(np.unique(y_test_pred, return_counts=True))

print(classification_report(y_test, y_test_pred))
j_train = jaccard_score(y_train, y_train_pred, average=None)
j_test = jaccard_score(y_test, y_test_pred, average=None)
print('Jaccard train', j_train)
print('Jaccard test', j_test)
print(np.mean(j_test))
print(np.std(j_test))
```
